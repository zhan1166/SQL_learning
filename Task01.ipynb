## 1.2 初识SQL

数据库中存储的表结构类似于excel中的行和列，在数据库中，行称为记录，它相当于一条记录，列称为字段，它代表了表中存储的数据项目

#### 根据对 RDBMS 赋予的指令种类的不同，SQL 语句可以分为以下三类:
    1) DDL（Data Definition Language，数据定义语言） 用来创建或者删除存储数据用的数据库以及数据库中的表等对象
    2) DML（Data Manipulation Language，数据操纵语言） 用来查询或者变更表中的记录
    3) DCL（Data Control Language，数据控制语言） 用来确认或者取消对数据库中的数据进行的变更以及对 RDBMS 的用户是否有权限操作数据库中的对象（数据库表等）进行设定

    DDL: CREATE, DROP, ALTER

    DML: SELECT, INSERT, UPDATE, DELETE

    DCL: COMMIT, ROLLBACK, GRANT, REVOKE

### 1.2.1 SQL 基本书写规则
    1) SQL语句要以;结尾
    2) SQL 不区分关键字的大小写，但是插入到表中的数据是区分大小写的
    3) win 系统默认不区分表名及字段名的大小写
    4) linux / mac 默认严格区分表名及字段名的大小写
    5) 常数的书写方式是固定的
      'abc', 1234, '26 Jan 2010', '10/01/26', '2010-01-26'......
    6) 单词需要用半角空格或者换行来分隔

### 1.2.2 数据类型的指定
    四种最基本的数据类型：

    INTEGER型 - 用来指定存储整数的列的数据类型（数字型），不能存储小数。

    CHAR型 - 用来存储定长字符串，当列中存储的字符串长度达不到最大长度的时候，使用半角空格进行补足，由于会浪费存储空间，所以一般不使用。

    VARCHAR型 - 用来存储可变长度字符串，定长字符串在字符数未达到最大长度时会用半角空格补足，但可变长字符串不同，即使字符数未达到最大长度，也不会用半角空格补足。

    DATE型 - 用来指定存储日期（年月日）的列的数据类型（日期型）。

### 1.2.3 约束条件
    约束是除了数据类型之外，对列中存储的数据进行限制或者追加条件的功能。
    1) NOT NULL是非空约束，即该列必须输入数据。
    2) PRIMARY KEY是主键约束，代表该列是唯一值，可以通过该列取出特定的行的数据。
